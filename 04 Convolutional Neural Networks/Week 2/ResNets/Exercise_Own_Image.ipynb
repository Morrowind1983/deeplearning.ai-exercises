```python
import numpy as np
import matplotlib.pyplot as plt

from keras.preprocessing import image
from matplotlib.pyplot import imshow

%matplotlib inline
```

```python
img_paths = ['test_image1.jpg', 'test_image2.jpg', 'test_image3.jpg', 'test_image4.jpg']
```

```python
def predict(model, input_shape, preprocess_input, decode_predictions):
    model = model(input_shape = input_shape)
    
    for img_path in img_paths:
        img = image.load_img(img_path, target_size=(input_shape[0], input_shape[1]))
        imshow(img)
        plt.show()

        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)

        preds = model.predict(x)
        # decode the results into a list of tuples (class, description, probability)
        print('Predicted:', decode_predictions(preds, top=5)[0])
```

### ResNet50

```python
from keras.applications import resnet50
predict(resnet50.ResNet50, (224, 224, 3), resnet50.preprocess_input, resnet50.decode_predictions)
```

### VGG16

```python
from keras.applications import vgg16
predict(vgg16.VGG16, (224, 224, 3), vgg16.preprocess_input, vgg16.decode_predictions)
```

### VGG19

```python
from keras.applications import vgg19
predict(vgg19.VGG19, (224, 224, 3), vgg19.preprocess_input, vgg19.decode_predictions)
```
